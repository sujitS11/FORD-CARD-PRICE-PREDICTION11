# EDA

```python
x_encoded=pd.get_dummies(X,drop_first=True)
x_onehot=x_encoded.astype(int)
```

```python
x_onehot.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,f1_score,classification_report
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
scaler=StandardScaler()
x_scaler=scaler.fit_transform(x_onehot)
```

```python
x_scaler
```

```python
xx=pd.DataFrame(x_scaler)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaler,Y,test_size=0.33,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
r2=r2_score(y_test,y_pred)
```

```python
r2
```

```python
n=x_test.shape[0]
p=x_test.shape[1]
adjusted_r2=1-((1-r2)*(n-1))/(n-p-1)
print(adjusted_r2)
```

```python
x_test.shape
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("R:", r2)
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
```

```python
y_train_pred = model.predict(x_train)
print("Train R:", r2_score(y_train, y_train_pred))
print("mae =", mean_absolute_error(y_train, y_train_pred))
print("mse :",mean_squared_error(y_train, y_train_pred))
print("Train RMSE:", np.sqrt(mean_squared_error(y_train, y_train_pred)))
```

```python
model.score(x_test,y_test),model.score(x_train,y_train)
```

```python
import joblib
joblib.dump(model,'LR_car.pkl')
joblib.dump(scaler,'scaler.pkl')
joblib.dump(x_onehot.columns.tolist(),'columns.pkl')
print("PKL file generate successfull")
```
